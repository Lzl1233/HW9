### Q1.3: What is a copula and how do the two versions of the code implement it?

A **copula** is a statistical tool that allows us to model the *dependency structure* between random variables separately from their *marginal distributions*. According to **Sklar’s Theorem**, any multivariate joint distribution can be decomposed into its marginals and a copula that binds them together:

\[
F(x_1, x_2, ..., x_n) = C(F_1(x_1), F_2(x_2), ..., F_n(x_n))
\]

Where:
- \(F_i(x_i)\) are the marginal CDFs
- \(C(\cdot)\) is the copula function capturing the dependence

---

**Implementation differences**:

- **First version (in the notebook)**:
  - Transforms each marginal variable \( y_i \) to a standard normal variable using the inverse CDF of the standard normal (Gaussian copula approach).
  - The joint dependency is modeled using a latent multivariate normal distribution (`MvNormal`) with an estimated correlation matrix via `LKJCholeskyCov`.
  - This method is **joint modeling**, building the copula and marginals in one PyMC model.

- **Second version (two-pass approach)**:
  - **Step 1**: Separately fit each marginal distribution and transform the observed data into standard normal space:
    \[
    u_i = F_i(y_i), \quad z_i = \Phi^{-1}(u_i)
    \]
  - **Step 2**: Fit a multivariate normal model on the transformed \((z_1, z_2, ..., z_n)\) to infer the dependency (copula).
  - This is a **modular** and flexible approach, useful when marginals are easier to model individually.

---

### Q1.4: How can copulas be used to build arbitrary multivariate GLM regressions?

Copulas can be used in multivariate regression settings where:

- The response variables follow **different distributions** (e.g., Gamma, Poisson, Exponential).
- We want to model the **joint behavior** and capture their correlation.

**Approach**:
1. Fit separate **GLMs** for each response variable:
   \[
   y_i \sim \text{Dist}_i(\mu_i = g_i^{-1}(X_i \beta_i))
   \]
2. Transform the fitted distributions into copula space using:
   \[
   u_i = F_i(y_i), \quad z_i = \Phi^{-1}(u_i)
   \]
3. Model the dependency across responses via a copula (e.g., Gaussian copula using multivariate normal).

This allows us to build **multivariate models** that flexibly combine different marginal behaviors with a shared dependency structure. It's widely used in finance, insurance, epidemiology, and any domain involving heterogeneous outcomes.


```python
import numpy as np
import pandas as pd

np.random.seed(0)
n = 100
p = 10
m = 2

X = np.random.normal(0, 1, (n, p))
B_true = np.zeros((p, m))
B_true[[1, 3, 5], 0] = [2.0, -1.5, 1.0]
B_true[[2, 4], 1] = [1.5, -2.0]

Y = X @ B_true + np.random.normal(0, 1, (n, m))
import pymc as pm
import arviz as az

with pm.Model() as spike_and_slab_model:
    slab_sd = pm.HalfNormal("slab_sd", sigma=1.0)

    inclusion = pm.Bernoulli("inclusion", p=0.5, shape=(p, m))
    weights_raw = pm.Normal("weights_raw", mu=0, sigma=1, shape=(p, m))
    weights = pm.Deterministic("weights", weights_raw * inclusion * slab_sd)

    mu = pm.math.dot(X, weights)
    sigma = pm.HalfNormal("sigma", sigma=1.0, shape=m)
    Y_obs = pm.Normal("Y_obs", mu=mu, sigma=sigma, observed=Y)

    idata_spike = pm.sample(1000, tune=1000, target_accept=0.9, chains=2, random_seed=42)
with pm.Model() as normal_prior_model:
    weights = pm.Normal("weights", mu=0, sigma=1.0, shape=(p, m))
    mu = pm.math.dot(X, weights)
    sigma = pm.HalfNormal("sigma", sigma=1.0, shape=m)
    Y_obs = pm.Normal("Y_obs", mu=mu, sigma=sigma, observed=Y)

    idata_normal = pm.sample(1000, tune=1000, target_accept=0.9, chains=2, random_seed=42)
import matplotlib.pyplot as plt
az.plot_forest(idata_spike, var_names=["weights"], combined=True, r_hat=True)
plt.title("Spike and Slab Coefficient Posterior");
az.plot_forest(idata_normal, var_names=["weights"], combined=True, r_hat=True)
plt.title("Normal Prior Coefficient Posterior");
```

### Summary and Complexity Analysis

The **Spike-and-Slab** model allows for explicit variable selection by introducing binary inclusion variables. This results in posterior distributions that clearly separate relevant variables (with non-zero weight distributions) from irrelevant ones (with weight distributions concentrated around zero or completely excluded).

In contrast, the **Diffuse Normal Prior** (standard L2-style prior) applies a weak Gaussian penalty to all coefficients, treating all predictors as potentially useful. As a result, the posterior distributions of coefficients are more homogeneous, and the model lacks sparsity.

#### Computational Complexity:

- **Sample Size (n)**: As n increases, both models become slower due to more likelihood evaluations. However, convergence in the Spike-and-Slab model can be slower due to the presence of discrete latent variables (Bernoulli inclusion indicators).

- **Response Dimension (m)**: The Spike-and-Slab model must estimate inclusion indicators and separate weights for each response variable, leading to higher memory and compute costs compared to the Normal Prior model.

- **Number of Predictors (p)**: In the Spike-and-Slab model, both the number of weights and inclusion indicators scale with p. This significantly increases model complexity compared to the Normal Prior model, which only needs to sample weights.

#### Conclusion:

While the Spike-and-Slab model provides stronger sparsity and interpretability, it is computationally more expensive. It is best suited for problems where variable selection is crucial and the dimensionality is moderate. The Diffuse Normal model, being simpler and faster, is a reasonable alternative when interpretability is less critical or when working with high-dimensional data.

### Comparison: Horseshoe Prior vs Spike-and-Slab

Both the **Horseshoe** and **Spike-and-Slab** priors aim to induce sparsity and perform variable selection in high-dimensional regression models. However, their mechanisms and properties differ:

- **Horseshoe Prior**: Uses continuous shrinkage based on a global scale (\( \tau \)) and local scales (\( \lambda_i \)). It allows large coefficients to escape shrinkage while strongly shrinking irrelevant ones. This results in smooth and adaptive regularization without hard thresholding.

- **Spike-and-Slab Prior**: Introduces binary inclusion indicators to explicitly turn off some coefficients. It produces more discrete posterior distributions and can be more interpretable, but introduces combinatorial complexity and convergence challenges.

#### Visual Comparison:
- The **Horseshoe** prior produces a continuous posterior with heavy tails (some weights can grow large).
- The **Spike-and-Slab** prior yields a bimodal posterior, with some weights tightly centered near zero and others clearly separated.

#### Practical Considerations:
- Horseshoe is easier to scale in MCMC because it avoids discrete variables.
- Spike-and-Slab may offer better variable exclusion for interpretation but is computationally more intensive.

In practice, the Horseshoe prior is often preferred for scalable sparse Bayesian regression, while Spike-and-Slab is used when strict variable selection is needed.
